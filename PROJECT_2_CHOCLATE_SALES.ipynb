# Task 5: Outlier Detection

* Detect Outliers Using a boxplot

```python
num_cols = ["Discount_Pct","Price_per_Box","Marketing_Spend",
            "Boxes_Shipped"]
for col in num_cols:
  sns.boxplot(x=df[col])
  plt.title(f'Boxplot of {col}')
  plt.show()
```

* Handle Outlier using the IQR

```python
for col in num_cols:
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower = Q1 - 1.5*IQR
  upper = Q3 + 1.5*IQR
  outliers = df = df[(df[col] >= lower) & (df[col] <= upper)]
  print(f"{col}:{len(outliers)}outliers")
```

* Handle Outlier Using Z-Score

```python
from scipy.stats import zscore
for col in num_cols:
  z_scores = zscore(df[col])
  outliers = df[np.abs(z_scores) > 3]
  print(f"{col}:{len(outliers)} outliers")
  df_clean = df[np.abs(z_scores) <= 3]
```

# Task 6: Univariate Analysis

# Task 7: Bivariate Analysis

* Analyze relationship between two variables using appropriate visualizations

```python
df['Amount']= pd.to_numeric(df['Amount'],errors='coerce')
plt.scatter(df['Boxes_Shipped'],df['Amount'])
plt.xlabel('Boxes_Shipped')
plt.ylabel('Amount')
plt.title('Relationship Between Boxes_Shipped vs Amount')
plt.show()
```

```python
df['Marketing_Spend']= pd.to_numeric(df['Marketing_Spend'],errors='coerce')
plt.scatter(df['Price_per_Box'],df['Marketing_Spend'])
plt.xlabel('Marketing_Spend')
plt.ylabel('Price_per_Box')
plt.title('Relationship Between Marketing_Spend vs Price_per_Box')
plt.show()
```

# Task 8: Correlation Heatmap

* Create a correlation heatmap

```python
df["amount"]= pd.to_numeric(df["Amount"],errors='coerce')
num_df = df.select_dtypes(include=["int64","float64"])
corr_matrix = num_df.corr()
sns.heatmap(corr_matrix,
            annot=True,
            cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

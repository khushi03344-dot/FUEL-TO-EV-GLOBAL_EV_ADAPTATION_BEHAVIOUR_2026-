```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)

y1 = x
y2 = x**2
y3 = np.sqrt(x) * 3
y4 = np.log(x + 1) * 5
y5 = np.sin(x) * 5

fig, axs = plt.subplots(3, 2, figsize=(12, 10))
axs = axs.flatten()


axs[0].plot(x, y1, color='red', marker='o', linewidth=2)
axs[0].set_title('Line Graph 1')
axs[0].grid(True)


axs[1].plot(x, y2, color='blue', marker='s', linewidth=2)
axs[1].set_title('Line Graph 2')
axs[1].grid(True)


axs[2].plot(x, y3, color='green', marker='^', linewidth=2)
axs[2].set_title('Line Graph 3')
axs[2].grid(True)


axs[3].plot(x, y4, color='orange', marker='D', linewidth=2)
axs[3].set_title('Line Graph 4')
axs[3].grid(True)

axs[4].plot(x, y5, color='purple', marker='*', linewidth=2)
axs[4].set_title('Line Graph 5')
axs[4].grid(True)


axs[5].axis('off')


plt.tight_layout()


plt.show()
```

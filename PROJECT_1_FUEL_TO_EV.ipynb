```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/global_ev_adoption_behavior_2026.csv")
sns.set_style("whitegrid")

```

**1 : Plot the distribution of age**

```python
plt.hist(df['age'])
plt.show()
sns.histplot(df['age'])
plt.show()
```

**2. Display the count of Educational Levels**

```python
df['education_level'].value_counts().plot(kind='bar')
plt.show()
sns.countplot(x='education_level', data=df)
plt.show()
```

**3: Show the distribution of Annual income**

```python
plt.hist(df['annual_income'])
plt.show()
sns.histplot(df['annual_income'],kde=True)
plt.show()
```

**4: Create a Pie Chart of city types**

```python
city = df['city_type'].value_counts()
plt.pie(city, labels=city.index, autopct='%1.1f%%')
plt.show()
```

**5: Compare Average Fuel Expense by City Type**

```python
avg = df.groupby('city_type')['fuel_expense_per_month'].mean()
plt.bar(avg.index, avg.values)
plt.show()
sns.barplot(data=df,
            x='city_type',
            y='fuel_expense_per_month')
plt.show()
```

**6: Plot Monthly Charging Cost Distribution**

```python
plt.hist(df['monthly_charging_cost'])
plt.show()
sns.histplot(df['monthly_charging_cost'])
plt.show()
```

**7: Plot  Current Vehicle type counts**

```python
vehicle = df['current_vehicle_type'].value_counts()
plt.bar(vehicle.index, vehicle.values)
plt.show()
sns.countplot(data=df,x='current_vehicle_type')
plt.show()
```

**8: Relationship between Fuel and Income Expense**

```python
sns.scatterplot(
    data=df,
    x='annual_income',
    y='fuel_expense_per_month'
)
plt.title("Income vs Fuel Expense")
plt.show()
```

**9: Charging Cost Trend by Vehicle Age**

---



```python

avg_cost = df.groupby('vehicle_age_years', as_index=False)['monthly_charging_cost'].mean()
plt.figure(figsize=(8,5))
sns.lineplot(
    data=avg_cost,
    x='vehicle_age_years',
    y='monthly_charging_cost',
    marker='o'
)
plt.title("Charging Cost Trend by Vehicle Age")
plt.xlabel("Vehicle Age (Years)")
plt.ylabel("Average Monthly Charging Cost")

plt.show()
```

**10: Create Multiple Subplots for Age and Annual Income**

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))
sns.histplot(df['age'], bins=20, color='skyblue', ax=ax[0])
ax[0].set_title("Age Distribution")
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Count")
sns.histplot(df['annual_income'], bins=20, color='orange', ax=ax[1])
ax[1].set_title("Annual Income Distribution")
ax[1].set_xlabel("Annual Income")
ax[1].set_ylabel("Count")
plt.tight_layout()
plt.show()
```

**11: Pairplot of selected features**

```python
sns.pairplot(df[['age',
                 'annual_income',
                 'daily_commute_km',
                 'monthly_charging_cost']])

plt.show()
```

**12: Correlation Heatmap**

```python
plt.figure(figsize=(12,8))

corr = df.corr(numeric_only=True)

sns.heatmap(corr,
            annot=True,
            cmap='coolwarm')

plt.show()
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("Europe Sales Records (4).csv")

# Total Revenue by Country
country_revenue = df.groupby("Country")["Total Revenue"].sum()

print("Total Revenue Generated by Each Country:")
print(country_revenue)

# Top 10 Countries by Revenue
top_10_countries = country_revenue.sort_values(ascending=False).head(10)

print("\nTop 10 Countries with Highest Revenue:")
print(top_10_countries)
```

```python
top_10_countries = country_revenue.sort_values(ascending=False).head(10)

print(top_10_countries)
```

```python
import pandas as pd
country_profit = df.groupby('Country')['Total Profit'].sum().sort_values(ascending=False)

print("Total Profit for each country:")
print(country_profit)
```

```python
top_5_profit = country_profit.head(5)

print("\nTop 5 countries with highest profit:")
print(top_5_profit)
```

```python
bottom_5_profit = country_profit.sort_values().head(5)

print("\nBottom 5 countries with lowest profit:")
print(bottom_5_profit)
```

```python
units_sold = df.groupby('Item Type')['Units Sold'].sum().sort_values(ascending=False)

print("Total Units Sold by Item Type:")
print(units_sold)
```

```python
revenue_by_item = df.groupby('Item Type')['Total Revenue'].sum()

highest_revenue_item = revenue_by_item.idxmax()
highest_revenue = revenue_by_item.max()

print("Item Type with Highest Revenue:")
print(highest_revenue_item, highest_revenue)
```

```python
profit_by_item = df.groupby('Item Type')['Total Profit'].sum()

highest_profit_item = profit_by_item.idxmax()
highest_profit = profit_by_item.max()

print("Item Type with Highest Profit:")
print(highest_profit_item, highest_profit)
```

```python
revenue_by_channel = df.groupby('Sales Channel')['Total Revenue'].sum()

print("Total Revenue by Sales Channel:")
print(revenue_by_channel)
```

```python
profit_by_channel = df.groupby('Sales Channel')['Total Profit'].sum()

print("Total Profit by Sales Channel:")
print(profit_by_channel)
```

```python
channel_summary = df.groupby('Sales Channel').agg({
    'Total Revenue': 'sum',
    'Total Profit': 'sum'
})
print(channel_summary)

best_revenue_channel = channel_summary['Total Revenue'].idxmax()
best_profit_channel = channel_summary['Total Profit'].idxmax()

print("Highest Revenue Channel:", best_revenue_channel)
print("Highest Profit Channel:", best_profit_channel)
```

```python
order_count = df['Order Priority'].value_counts()

avg_profit = df.groupby('Order Priority')['Total Profit'].mean()

highest_priority = avg_profit.idxmax()

print("Order Counts:")
print(order_count)

print("\nAverage Profit by Priority:")
print(avg_profit)

print("\nHighest Average Profit Priority:")
print(highest_priority)
```

```python
correlation = df['Total Revenue'].corr(df['Total Profit'])

print("Correlation between Total Revenue and Total Profit:", correlation)
```

if the corelation is strongly positive then hgher revenue generally leads to high profit

```python
import pandas as pd

df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_revenue = df.groupby(df['Order Date'].dt.to_period('M'))['Total Revenue'].sum()

print("Monthly Revenue:")
print(monthly_revenue)
highest_month = monthly_revenue.idxmax()
highest_value = monthly_revenue.max()

print("Highest Revenue Month:", highest_month)
print("Revenue:", highest_value)
lowest_month = monthly_revenue.idxmin()
lowest_value = monthly_revenue.min()

print("Lowest Revenue Month:", lowest_month)
print("Revenue:", lowest_value)
```
